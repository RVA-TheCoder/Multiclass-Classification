#### (vi) Fetching numerical and categorical columns

```python
def num_cat_col_list(DF) : 
    
    from pandas.api.types import is_string_dtype, is_numeric_dtype

    
    num_list = []
    cat_list = []

    for column in DF:

        if is_numeric_dtype(DF[column]):
            num_list.append(column)

        elif is_string_dtype(DF[column]):
            cat_list.append(column)

    return (num_list , cat_list)
```

```python
num_list , cat_list = num_cat_col_list(telecom)

print("\33[1mNumerical column :\33[0m ",num_list)
print()
print("\33[1mCategorical column :\33[0m ",cat_list)
```

## <span style = 'background :lightblue'>STEP 3 : UNIVARIATE ANALYSIS</span>

### (i) Histogram-kde  for Numerical  attributes.

```python
def hist_kde_plot(DF , num_list):
    
    total_size1 = len(num_list) 

    n_cols1 = 2
    n_rows1 = int(np.ceil(total_size1/n_cols1))

    fig1,axes1 = plt.subplots(nrows = n_rows1 ,ncols = n_cols1 , figsize = (12,8))

    i = 0
    j = 0

    fig1.suptitle("HISTOGRAM-Kde Plot \n", fontweight="bold" ,fontsize=20)


    for col in num_list:


        axes2 = sns.histplot(ax = axes1[i,j], data = DF , x = col ,kde =  True ,bins = 20)
        fig1.tight_layout()

        axes2.set_title(f"{col}",fontweight="bold" ,fontsize=12)

        if j == n_cols1 -1:

            i +=1
            j = 0

        else :
            j+=1  
```

```python
hist_kde_plot(DF = telecom, num_list=num_list)
```

### (ii) countplot for  categorical attributes 

```python
def countplot_cat_col(DF,cat_list):
    
    total_size2 = len(cat_list)

    n_cols2 = 2
    n_rows2 = int(np.ceil(total_size2/n_cols2))

    fig3,axes3 = plt.subplots(nrows = n_rows2 ,ncols = n_cols2 , figsize = (12,20))
    fig3.tight_layout()

    i = 0
    j = 0
    
    fig3.suptitle("COUNTPLOTS \n", fontweight="bold" ,fontsize=20)

    for col in cat_list:

        fig3.tight_layout()

        axes4 = sns.countplot(ax = axes3[i,j], data = DF , x = col ,hue = col )

        axes4.set(xticklabels=[])
        axes4.set(xlabel=None)
        axes4.set_title(f"{col}",fontweight="bold" ,fontsize=12)

        axes4.set(ylabel='Frequency')
        axes4.grid(visible=True ,axis ='y')
        
        if j == n_cols2 -1 :
        
            i +=1
            j = 0
        
        else :
            j += 1 
            
```

```python
countplot_cat_col(DF = telecom , cat_list=cat_list)
```

### (iii) countplot for  categorical attributes with hue

```python
def countplt_with_hue(DF,cat_list_hue):
    
    n_cols = 2
    n_rows = int(np.ceil(len(cat_list_hue)/ n_cols))

    
    fig5, axes5 = plt.subplots(nrows=n_rows, ncols=n_cols , figsize = (14,22),)
    fig5.suptitle("COUNTPLOTS WITH HUE : CHURN\n", fontweight = 'bold' , fontsize = 20)
    
    i = 0
    j = 0

    for comb in cat_list_hue:

        fig5.tight_layout()

        axes6 = sns.countplot(ax =axes5[i,j],data = DF , x = comb , hue = 'Churn')

        axes6.set_title(comb,fontweight="bold" ,fontsize=14)
        axes6.set(ylabel = 'COUNT', xlabel = " ")
        axes6.set_xticklabels(labels=list(DF[comb].unique()), rotation=30)
        axes6.grid(visible=True ,axis ='y')

        if j == n_cols-1:
            i = i + 1
            j = 0

        else:
            j = j + 1
```

```python
catlist_hue = ['SeniorCitizen','gender' ,'Partner' ,'Dependents', 'PhoneService' , 'InternetService' , 'OnlineSecurity' , 'TechSupport','StreamingTV' , 'Contract' , 'PaperlessBilling' , 'PaymentMethod']
```

```python
countplt_with_hue(DF = telecom,cat_list_hue = catlist_hue)
```

### (i) Numerical vs. Numerical â†’ heat map or pairplot

```python
def correlation_map(DF):
    
    correlation = round(DF.corr(numeric_only =True),3)
    
    plt.figure(figsize = (4,3) ,dpi = 120 , layout = 'tight' )
    sns.heatmap(correlation,cmap = 'GnBu' , annot = True)

    plt.show()  
```

```python
correlation_map(DF = telecom)
```

### (ii) categorical vs. categorical 

#### Focussing on Churned customer

```python
churn_df = telecom.loc[telecom['Churn'] == 'Yes']
```

```python
churn_df.shape
```

#### Custom Function for Discrete combinations

```python
list1 = ['SeniorCitizen','gender' ,'Partner' , 'StreamingTV' , 'Contract'  , 'PaymentMethod']
```

```python
def catcomblist(listobject) : 
      
    list2 = []

    list3 = listobject.copy()

    x = len(listobject)


    for ele in listobject :

        #print("Element to be removed : ",ele)

        if  x > 1 :

            for i in range(1,len(list3)) :


                list2.append([ele,list3[i]])               

            list3.remove(ele)
            #print("list3 :" , list3)

            x -=1

    return list2
```

```python
# cat_comb_list = catcomblist(listobject = list1)
# cat_comb_list
```

```python
def countplot_churned_cust(DF ,comb_list ) : 
    
    n_cols = 2
    n_rows =  int(np.ceil(len(comb_list)/n_cols))


    fig11 ,axes11 = plt.subplots(nrows= n_rows  , ncols =n_cols , figsize = (15,24) )

    i = 0
    j = 0

    fig11.suptitle("CHURNED CUSTOMERS\n", fontweight="bold" ,fontsize=20)

    for cat in comb_list:
        
        # print(cat[0])
        # print(cat[1])
        fig11.tight_layout()


        hue_cat = cat[0]
        col = cat[1]

        axes12 = sns.countplot(ax = axes11[i ,j], data = DF , x = col ,
                      order = DF[col].value_counts().index , 
                      hue = hue_cat , 
                      palette = 'bright')

        axes12.set_title(f"{col} vs {hue_cat}",fontweight="bold" ,fontsize=14)
        axes12.grid(visible=True ,axis ='y')


        if j == n_cols -1:

            i += 1
            j = 0 

        else :

            j +=1
```

```python
countplot_churned_cust(DF = churn_df  ,comb_list  = catcomblist(listobject = list1))
```
